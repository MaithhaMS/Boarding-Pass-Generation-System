```python

class Passenger:
    def __init__(self, name, age, nationality, idNumber, passportNumber):
        self.name = name
        self.age = age
        self.nationality = nationality
        self.idNumber = idNumber
        self.passportNumber = passportNumber

    def get_name(self):
        return self.name

    def set_name(self, name):
        self.name = name

    def get_age(self):
        return self.age

    def set_age(self, age):
        self.age = age

    def get_nationality(self):
        return self.nationality

    def set_nationality(self, nationality):
        self.nationality = nationality

    def get_idNumber(self):
        return self.idNumber

    def set_idNumber(self, idNumber):
        self.idNumber = idNumber

    def get_passportNumber(self):
        return self.passportNumber

    def set_passportNumber(self, passportNumber):
        self.passportNumber = passportNumber

    def displayInfo(self):
        print("Passenger Information:")
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
        print(f"Nationality: {self.nationality}")
        print(f"Id Number: {self.idNumber}")
        print(f"Passport Number: {self.passportNumber}\n")



class Flight:
    def __init__(self, flight_number, origin, destination, date_time, gate):
        self.flight_number = flight_number
        self.origin = origin
        self.destination = destination
        self.date_time = date_time
        self.gate = gate

    def get_flight_number(self):
        return self.flight_number

    def set_flight_number(self, flight_number):
        self.flight_number = flight_number

    def get_origin(self):
        return self.origin

    def set_origin(self, origin):
        self.origin = origin

    def get_destination(self):
        return self.destination

    def set_destination(self, destination):
        self.destination = destination

    def get_date_time(self):
        return self.date_time

    def set_date_time(self, date_time):
        self.date_time = date_time

    def get_gate(self):
        return self.gate

    def set_gate(self, gate):
        self.gate = gate

    def displayInfo(self):
        print("Flight Information:")
        print(f"Flight Number: {self.flight_number}")
        print(f"Origin: {self.origin}")
        print(f"Destination: {self.destination}")
        print(f"Date and Time: {self.date_time}")
        print(f"Gate: {self.gate}\n")


class BoardingPass:
    def __init__(self, passenger, flight_class, boarding_time, seat_number, electronic_ticket):
        self.passenger = passenger
        self.flight_class = flight_class
        self.boarding_time = boarding_time
        self.seat_number = seat_number
        self.electronic_ticket = electronic_ticket

    def get_passenger(self):
        return self.passenger

    def set_passenger(self, passenger):
        self.passenger = passenger

    def get_flight_class(self):
        return self.flight_class

    def set_flight_class(self, flight_class):
        self.flight_class = flight_class

    def get_boarding_time(self):
        return self.boarding_time

    def set_boarding_time(self, boarding_time):
        self.boarding_time = boarding_time

    def get_seat_number(self):
        return self.seat_number

    def set_seat_number(self, seat_number):
        self.seat_number = seat_number

    def get_electronic_ticket(self):
        return self.electronic_ticket

    def set_electronic_ticket(self, electronic_ticket):
        self.electronic_ticket = electronic_ticket

    def displayInfo(self):
        print("Boarding Pass Information:")
        print(f"Passenger: {self.passenger.name}")
        print(f"Flight Class: {self.flight_class}")
        print(f"Boarding Time: {self.boarding_time}")
        print(f"Seat Number: {self.seat_number}")
        print(f"Electronic Ticket : {self.electronic_ticket}\n")



class BoardingPassSystem:
    def __init__(self):
        self.boarding_pass_list = []

    def generate_boarding_pass(self, passenger, flight_class, boarding_till, seat_number, electronic_ticket):
        boarding_pass = BoardingPass(passenger, flight_class, boarding_till, seat_number, electronic_ticket)
        self.boarding_pass_list.append(boarding_pass)

    def display_boarding_pass_details(self):
        for boarding_pass in self.boarding_pass_list:
            boarding_pass.displayInfo()

    def update_boarding_information(self, boarding_pass, new_gate, new_boarding_time):
        # Functionality to update boarding information (not implemented in this example)
        pass


# Create a Passenger object
passenger = Passenger("JAMES SMITH", "25", "UK", "1234567", "JS1234567")

# Create a Flight object
flight = Flight("NA4321", "Chicago", "New York", "2020-12-06 at 11:40", "03")

# Create a BoardingPass object
boarding_pass = BoardingPass(passenger, "First Class", "11:20", "09A", "629")

# Create a BoardingPassSystem object
boarding_pass_system = BoardingPassSystem()

# Generate a boarding pass and add it to the boarding pass system
boarding_pass_system.generate_boarding_pass(passenger, "First Class", "Till 11:20", "09A", "629")

# Display boarding pass details
boarding_pass_system.display_boarding_pass_details()

passenger.displayInfo()
flight.displayInfo()
boarding_pass.displayInfo()
```
